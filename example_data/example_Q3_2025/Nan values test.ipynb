```python
# ROBUST STREAMING (BALANCE + CLEAN)
# 1. Counts total rows.
# 2. Sets probability for balancing.
# 3. CLEANS data (NaN -> -1) row-by-row.
# 4. Saves the final small, clean file.

import csv
import random
from pathlib import Path

# --- CONFIGURATION ---
# UPDATE THIS PATH to your real 40GB file location
csv_path = Path('/work/shibberu/share/MA384_Data_Mining_Projects_Winter_2025-26/SMART_failure_prediction/S.M.A.R.T/example_data/example_Q3_2025/balanced_dataset_final.csv') 
out_path = Path('balanced_clean_dataset.csv')

# Use Global Probability for the 40GB run (e.g., 0.00035)
# Use Target Logic ONLY for the single-file test
target_successes = 20_000   
possible_targets = ['failure','failed','target','label','y','is_failed']
random_seed = 42
progress_interval = 500_000

random.seed(random_seed)

# --- HELPER FUNCTION ---
def find_target_idx(header):
    lower = [h.strip().lower() for h in header]
    for t in possible_targets:
        if t in lower:
            return lower.index(t)
    for i, h in enumerate(lower):
        if 'failure' in h or 'fail' == h:
            return i
    return None

if not csv_path.exists():
    raise FileNotFoundError(f'CSV not found: {csv_path}')

# --- PASS 1: COUNT TOTAL ROWS ---
print("Pass 1: Counting total rows...")
with csv_path.open(newline='', encoding='utf-8', errors='replace') as f:
    total_rows = sum(1 for _ in f) - 1

print(f"Total rows: {total_rows}")

# --- CALCULATION ---
# (Switch this to keep_prob = 0.00035 for the full 40GB run)
if total_rows <= target_successes:
    keep_prob = 1.0
else:
    keep_prob = target_successes / total_rows

print(f"Sampling Probability: {keep_prob:.6f}")

# --- PASS 2: STREAM, CLEAN, FILTER ---
print("Pass 2: Cleaning and Filtering...")
kept_failures = 0
kept_successes = 0
rows_seen = 0

with csv_path.open(newline='', encoding='utf-8', errors='replace') as fin, \
     out_path.open('w', newline='', encoding='utf-8') as fout:
    
    reader = csv.reader(fin)
    writer = csv.writer(fout)
    
    # Header
    header = next(reader)
    writer.writerow(header)
    
    target_idx = find_target_idx(header)
    if target_idx is None:
        raise ValueError("Target column missing.")

    for row in reader:
        rows_seen += 1
        
        if len(row) <= target_idx:
            continue
            
        val = row[target_idx].strip()
        
        # Check Failure Status
        try:
            v = int(float(val))
        except:
            v = 1 if val.lower() in ('true','t','yes','y','1') else 0
            
        # DECISION: Do we keep this row?
        keep_it = False
        if v == 1:
            keep_it = True # Keep all failures
        elif random.random() < keep_prob:
            keep_it = True # Random sample successes
            
        if keep_it:
            # --- CLEANING STEP (NEW) ---
            # Replace empty strings (NaNs) with "-1"
            clean_row = [x if x.strip() != '' else '-1' for x in row]
            
            writer.writerow(clean_row)
            
            if v == 1:
                kept_failures += 1
            else:
                kept_successes += 1
        
        if rows_seen % progress_interval == 0:
            print(f"Processed {rows_seen} rows...")

print("--- DONE ---")
print(f"Final Data: {kept_failures} Failures, {kept_successes} Successes")
print(f"Saved clean file to: {out_path.resolve()}")
```
